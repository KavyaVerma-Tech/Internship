```python
def factorial(num):
  factorial = 1
  if num == 0:
    return 1
  for i in range(1 , num + 1):
    factorial = factorial * i
  return factorial
num = int(input("Enter a number: "))
print("Factorial of", num, "is", factorial(num))
```

```python
def multiplication(num):
  for i in range(1,10+1):
    print(num," * ",i, " = " ,num * i)
num = int(input("Enter a number: "))
print("Multiplication Table of", num, "is:")
multiplication(num)
```

```python
def prime(num):
  if num < 2:
    print("Not Prime")
  for i in range(2,num):
    if num % i == 0:
      print("Not Prime")
      break
  else:
      print("Prime")

num = int(input("Enter a number: "))
prime(num)
```

```python
def largest(l):
  largest = l[0]
  for i in range(1,len(l)):
    if l[i] > largest:
      largest = l[i]
  return largest
l = []
n = int(input("ënter no of elements in list: "))
for i in range(0,n):
  num=int(input("enter the number"))
  l.append(num)
print(l)
largest(l)
```
